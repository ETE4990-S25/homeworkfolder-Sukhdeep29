```python
import multiprocessing
import threading
import asyncio
import time

def is_prime(n):
    if n <=1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i ==0:
            return False
    return True

def highest_prime(time_limit=180):
    initial_time= time.time()
    num= 2
    prime_num=2
    while time.time() - initial_time < time_limit:
        if is_prime(num):
            prime_num= num
        num = num+1
    return prime_num

def fibonacci(n):
    if n<=1:
        return n
    a=0
    b=1
    for i in range(2, n+1):
        temp= a+b
        a=b
        b=temp
    return b  


async def factorial(n):
    result= 1
    for i in range(2, n+1):
        result = result*i
    return result  
    

def run_fibonacci(n):
    print("Fibonacci(" +str(n) + ")=" + str(fibonacci(n)))


def run_factorial(n):
    loop = asyncio.get_event_loop()
    result= loop.run_until_complete(factorial(n))
    print("Factorial(" + str(n) + ")=" +str(result))


def main():
    with multiprocessing.Pool(processes=1) as pool:
        prime_num= pool.apply(highest_prime)
    print("Highest Prime is: " + str(prime_num))


    fib_thread = threading.Thread(target= run_fibonacci, args= (prime_num,))
    fib_thread.start()

    fac_thread= threading.Thread(target=run_factorial, args=(prime_num,))
    fac_thread.start()


    fib_thread.join()
    fac_thread.join()


if __name__== "__main__": ####Learned about using this especially running it on windows. (from the internet)
    ###asyncio.run(main())  ###Didn't work. Got some help from the internet and found it can't be called from a running event loop. 
    try: 
        main()
    except RuntimeError:
        print("Event loop is already running.")
        main() 
    
```

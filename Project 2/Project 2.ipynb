```python
import multiprocessing
import threading
import asyncio
import time

##def is_prime(n):
    #if n <=1:
        #return False
    #for i in range(2, int(n**0.5)+1):
        #if n%i ==0:
            #return False
    #return True
def sieve_of_eratosthenes(limit):
    primes= [True] * (limit+1)
    primes[0]= primes[1]= False
    for i in range(2, int(limit**0.5)+1):
        if primes[i]:
            for j in range(i*i, limit +1, i):
                primes[j]= False
    return[i for i, prime in enumerate(primes)if prime]


def highest_prime(time_limit=180):
    start_time= time.time()
    limit= 10000000

    while time.time() - start_time < time_limit:
        primes= sieve_of_eratosthenes(limit)
        if primes:
            prime_num= primes[-1]
            return prime_num
        limit = limit*2
    return 2

def fibonacci(n):
    if n<=1:
        return n
    a,b = 0,1
    for _ in range(2, n+1):
        a, b = b, a+b
    return b  


def factorial(n):
    result= 1
    for i in range(2, n+1):
        result = result*i
    return result  
    

async def async_fib(n):
    result= await asyncio.to_thread(fibonacci, n)
    print(f"Fibonacci ({n})= {result}")

async def async_fac(n):
    result= await asyncio.to_thread(factorial, n)
    print(f"Factorial({n})= {result}")
    
def multi_core():
    with multiprocessing.Pool(processes=1) as pool:
        prime_num= pool.apply(highest_prime)
    print(f"Highest Prime using Multi Core: {prime_num}")

    fib_thread= threading.Thread(target=asyncio.run, args=(async_fib(prime_num),))
    fac_thread= threading.Thread(target=asyncio.run, args=(async_fac(prime_num),))

    fib_thread.start()
    fac_thread.start()

    fib_thread.join()
    fac_thread.join()

def using_thread():
    prime_num= highest_prime()
    print(f"Highest prime using threading: {prime_num}")

    fib_thread= threading.Thread(target=asyncio.run, args=(async_fib(prime_num),))
    fac_thread= threading.Thread(target=asyncio.run, args=(async_fac(prime_num),))
    
    fib_thread.start()
    fac_thread.start()

    fib_thread.join()
    fac_thread.join()


def main():
    start_time =time.time()

    print("Running Multi Core:")
    multi_core()

    print("\n" + "=" *30+ "\n")

    print("Running Threading:")
    using_thread()

    end_time= time.time()
    print(f"\n Total run time: {end_time - start_time: 0.2f} seconds")


if __name__== "__main__": ####Learned about using this especially running it on windows. (from the internet)
    ###asyncio.run(main())  ###Didn't work. Got some help from the internet and found it can't be called from a running event loop. 
    main() 
    
```
